```python
import math 
import numpy as np
import random
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt
```

```python
def f(x):
    return (-2 * math.cos(x) / math.pi) + (x / math.pi) - (2 * math.cos(3 * x) / (3 * math.pi))
```

```python
def generate_data_with_poisson_noise(scaling_factor, total_number_of_data_points):
    dataset = []
    while total_number_of_data_points > 0:
        x = round(random.uniform(-10, 10), 2)
        noise = scaling_factor * np.random.poisson(2)
        y = f(x) + noise
        dataset.append((x,y))
        total_number_of_data_points-=1
    return dataset
```

```python
def find_error(Y_true, Y_pred):
    n = len(Y_true)
    mean_square_error = 0
    for i in range (n):
       mean_square_error = mean_square_error + (Y_true[i] - Y_pred[i]) ** 2
    mean_square_error = mean_square_error / n
    return mean_square_error
```

```python
def polynomial_regression_sklearn(X, Y, degree):
    X = np.array(X)
    Y = np.array(Y)
    X = X.reshape(-1, 1)
    poly = PolynomialFeatures(degree=degree, include_bias=False)
    X_poly = poly.fit_transform(X)
    model = LinearRegression()
    model.fit(X_poly, Y)
    return np.insert(model.coef_, 0, model.intercept_)
```

```python
def polynomial_regressor_poisson(X, Y, degree):
    coefficients = polynomial_regression_sklearn(X, Y, degree)
    Y_pred = []
    n = len(X)
    for i in range (n):
        y_pred = 0
        for deg in range (degree+1):
            y_pred = y_pred + coefficients[deg] * pow(X[i], deg)
        Y_pred.append(y_pred)
    return Y_pred
```

```python
random.seed(65)
total_number_of_data_points = 50
scaling_factor = 0.1
dataset50 = generate_data_with_poisson_noise(scaling_factor, total_number_of_data_points)
X = [dataset_instance[0] for dataset_instance in dataset50]
Y = [dataset_instance[1] for dataset_instance in dataset50]
```

```python
empirical_errors = []
m = []
for i in range (1,21):
    Y_pred = polynomial_regressor_poisson(X, Y, i)
    empirical_error = find_error(Y, Y_pred)
    m.append(i)
    empirical_errors.append(empirical_error)

title = 'empirical error vs m'
fig, ax = plt.subplots()
ax.set_xlim(0, 21)
ax.set_ylim(0, 0.4)
plt.plot(m, empirical_errors, marker='o', linestyle='-', color='b', label='Data Points')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title(title)
plt.legend()  
plt.show()
```

```python
def sample_dataset(X, Y, no_of_samples):
    dataset_size = len(X)
    sampled_dataset = []
    while no_of_samples > 0:
        idx = random.randint(0, dataset_size - 1)
        sampled_dataset.append((X[idx], Y[idx]))
        no_of_samples-=1
    return sampled_dataset
```

```python
def calculate_mean(Y_pred):
    return list(np.array(Y_pred).mean(axis=0))
```

```python
def calculate_bias(Y, Y_pred_mean):
    return np.mean((Y - Y_pred_mean)**2)
```

```python
def calculate_variance(Y_pred):
    return np.mean(np.var(Y_pred, axis=0))
```

```python
biases = []
variances = []
m = []
print(X)
print(Y)
for i in range (1, 21):
    m.append(i)
    poly = PolynomialFeatures(degree = i, include_bias=False)
    X = np.array(X)
    Y = np.array(Y)
    X = X.reshape(-1, 1)
    X_poly = poly.fit_transform(X)
    no_of_sample = 200
    Y_pred = []
    Y_pred_mean = []
    for j in range (1, 101):
        sampled_dataset = sample_dataset(X, Y, no_of_sample)
        X_j = [dataset_instance[0] for dataset_instance in sampled_dataset]
        Y_j = [dataset_instance[1] for dataset_instance in sampled_dataset]
        X_j = np.array(X_j)
        Y_j = np.array(Y_j)
        X_j = X_j.reshape(-1, 1)
        poly = PolynomialFeatures(degree = i, include_bias=False)
        X_poly_j = poly.fit_transform(X_j)
        model = LinearRegression()
        model.fit(X_poly_j, Y_j)
        y_pred = model.predict(X_poly)
        Y_pred.append(y_pred)
    Y_pred_mean = calculate_mean(Y_pred)
    bias = calculate_bias(Y, Y_pred_mean)
    variance = calculate_variance(Y_pred)
    biases.append(bias)
    variances.append(variance)
print(biases)
print(variances)
title = 'bias and variances vs m'
fig, ax = plt.subplots()
ax.set_xlim(0, 21)
ax.set_ylim(0, 0.5)
plt.plot(m, biases, marker='o', linestyle='-', color='yellow', label='bias')
plt.plot(m, variances, marker='o', linestyle='-', color='green', label='variance')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title(title)
plt.legend()  
plt.show()
```
