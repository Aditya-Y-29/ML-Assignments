```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Generate some sample data
np.random.seed(0)
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Feature matrix (including intercept term)
y = np.array([2, 3, 4, 5, 6])  # Target vector

print(X, y)
# Polynomial features transformation
degree = 2  # Degree of the polynomial
poly_features = PolynomialFeatures(degree=degree)
X_poly = poly_features.fit_transform(X)

print(X_poly)

# Fit linear regression model
model = LinearRegression()
model.fit(X_poly, y)

print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
# Make predictions
X_new = np.linspace(-1, 1, 100).reshape(-1, 1)
X_new_poly = poly_features.transform(X_new)
y_new = model.predict(X_new_poly)

# Plot the results
plt.scatter(X, y, label='Original Data')
plt.plot(X_new, y_new, color='red', label='Polynomial Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Polynomial Regression')
plt.legend()
plt.show()
```
